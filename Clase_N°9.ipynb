##âœ…**Caso 1. Sistema de recomendaciÃ³n de PelÃ­culas**
---



* Un sistema de recomendaciÃ³n de pelÃ­culas podrÃ­a utilizar el Ã¡lgebra booleana para recomendar pelÃ­culas a un usuario que ha indicado que **le gustan las pelÃ­culas de acciÃ³n y ciencia ficciÃ³n.** El sistema podrÃ­a buscar pelÃ­culas que sean tanto de acciÃ³n como de ciencia ficciÃ³n, y luego recomendarlas al usuario.

âš¡**ExplicaciÃ³n:**

* La funciÃ³n ```recomendar_peliculas``` recibe un diccionario con las preferencias del usuario como entrada.

* La variable ```peliculas``` es un diccionario que mapea cada gÃ©nero con una lista de pelÃ­culas de ese gÃ©nero.

* La funciÃ³n recorre los gÃ©neros favoritos del usuario y agrega las pelÃ­culas de cada gÃ©nero a la lista de recomendaciones.

* Finalmente, se eliminan los duplicados de la lista y se retorna la lista de pelÃ­culas recomendadas.

ðŸ“‹ **APLICACIÃ“N EN PHYTON** ðŸ“‹

````python
def recomendar_peliculas(usuario):
  """
  FunciÃ³n que recomienda pelÃ­culas a un usuario en funciÃ³n de sus preferencias.

  ParÃ¡metros:
    usuario: Diccionario con las preferencias del usuario.
  Retorno:
    Lista de pelÃ­culas recomendadas.
  """

  peliculas = {
    "accion": ["Duna", "Demon Slayer", "El origen", "MÃ¡x", "Resistencia"],#
    "ciencia_ficcion": ["El origen", "La llegada", "Interestelar","The Cloverfield Paradox", "Time Trap"],
    "comedia": ["Wonka", "Ascenso", "Intensamente", "Pared con Pared", "A travÃ©s de tu Mirada"],
  }

  recomendaciones = []
  for genero in usuario["generos_favoritos"]:
    recomendaciones.extend(peliculas[genero])

  # Eliminar duplicados
  recomendaciones = set(recomendaciones)

  return list(recomendaciones)

usuario = {
  "nombre": "Diego",
  "generos_favoritos": ["accion", "ciencia_ficcion"]
}

recomendaciones = recomendar_peliculas(usuario)

print("PelÃ­culas recomendadas para", usuario["nombre"], ":", recomendaciones)

````

ðŸ¤” **Â¿DÃ³nde se aplica el Ã¡lgebra booleana en el caso anterior?**

En el ejemplo anterior, el Ã¡lgebra booleana se aplica en la siguiente parte:

    for genero in usuario["generos_favoritos"]:
    recomendaciones.extend(peliculas[genero])

En esta secciÃ³n, se recorre la lista de gÃ©neros favoritos del usuario (```usuario["generos_favoritos"]```). Para cada gÃ©nero, se utiliza la clave del gÃ©nero (```genero```) como Ã­ndice para acceder a la lista de pelÃ­culas de ese gÃ©nero en el diccionario ```peliculas```.

**AquÃ­ es donde se aplica el Ã¡lgebra booleana:**

* La clave del gÃ©nero (```genero```) se utiliza como condiciÃ³n booleana.
* Si el gÃ©nero estÃ¡ presente en la lista de gÃ©neros favoritos del usuario, la expresiÃ³n se evalÃºa como ```True```.
* Si el gÃ©nero no estÃ¡ presente en la lista de gÃ©neros favoritos del usuario, la expresiÃ³n se evalÃºa como ```False```.

##âœ…**Caso 2. ClasificaciÃ³n de textos y anÃ¡lisis de sentimientos**
---



* Un sistema de anÃ¡lisis de opiniones podrÃ­a utilizar el Ã¡lgebra booleana para determinar si un tweet es **positivo**, **negativo o neutral** sobre un producto en particular. El sistema podrÃ­a buscar palabras clave que se asocian con opiniones positivas o negativas, y luego determinar la opiniÃ³n general del tweet.

âš¡**ExplicaciÃ³n:**

* La funciÃ³n ```analizar_opinion``` recibe el texto del tweet y el nombre del producto como entrada.

* Las variables ```palabras_positivas``` y ```palabras_negativas``` contienen listas de palabras que se asocian con opiniones positivas y negativas, respectivamente.

* La funciÃ³n recorre las palabras del tweet y aumenta o disminuye la puntuaciÃ³n en funciÃ³n de si la palabra es positiva o negativa.

* Finalmente, se retorna la opiniÃ³n general del tweet en funciÃ³n de la puntuaciÃ³n:

        Si la puntuaciÃ³n es positiva, se retorna "Positivo".

        Si la puntuaciÃ³n es negativa, se retorna "Negativo".

        Si la puntuaciÃ³n es cero, se retorna "Neutral".

ðŸ“‹ **APLICACIÃ“N EN PHYTON** ðŸ“‹

````python
def analizar_opinion(tweet, producto):
  """
  FunciÃ³n que analiza la opiniÃ³n sobre un producto en un tweet.

  ParÃ¡metros:
    tweet: Texto del tweet.
    producto: Nombre del producto.

  Retorno:
    "Positivo", "Negativo" o "Neutral".
  """

  palabras_positivas = ["bueno", "excelente", "maravilloso", "me gusta", "me encanta", "lo quiero", "lo necesito"]
  palabras_negativas = ["malo", "terrible", "horrible", "no me gusta", "pÃ©simo", "desagradable", "mal estado"]

  puntuacion = 0
  for palabra in tweet.split():
    if palabra in palabras_positivas:
        puntuacion += 1
    elif palabra in palabras_negativas:
      puntuacion -= 1

  if puntuacion > 0:
    return "Positivo"
  elif puntuacion < 0:
    return "Negativo"
  else:
    return "Neutral"

tweet = "El nuevo telÃ©fono #SamsungGalaxyS23 es excelente! #Tecnologia #Smartphone"
producto = "Samsung Galaxy S23"

import re
#Utilizamos una expresiÃ³n regular para eliminar los sÃ­mbolos
texto_limpio = re.sub(r'[^\w\s]', '', tweet)

opinion = analizar_opinion(texto_limpio, producto)

print("OpiniÃ³n sobre", producto, ":", opinion)
````

* En Phyton, se podrÃ­a implementar un sistema experto para diagnosticar enfermedades respiratorias de la siguiente manera:

**1. Base de conocimiento:**

* Esta informaciÃ³n se puede almacenar en una estructura de datos como un diccionario o una lista de objetos.


**2. RecopilaciÃ³n de sÃ­ntomas**

* Se le pregunta al usuario sobre los sÃ­ntomas que estÃ¡ experimentando.
* Esta informaciÃ³n se puede recopilar a travÃ©s de una interfaz grÃ¡fica de usuario o mediante preguntas de consola.

**3. AnÃ¡lisis de sÃ­ntomas**

* Se utiliza la informaciÃ³n de la base de conocimiento y el Ã¡lgebra booleana para analizar los sÃ­ntomas del usuario y determinar la enfermedad mÃ¡s probable.

* Se pueden utilizar reglas booleanas para determinar si un conjunto de sÃ­ntomas es compatible con una enfermedad en particular.


**4. PresentaciÃ³n del diagnÃ³stico:**

* Se presenta al usuario el diagnÃ³stico mÃ¡s probable, junto con informaciÃ³n adicional sobre la enfermedad, como sus causas, tratamiento y prevenciÃ³n.

ðŸ¦ **EJEMPLOS DE APLICACIÃ“N DEL ÃLGEBRA BOOLEANA** ðŸ¦ 

**Regla 1:** Si el usuario tiene fiebre, tos y dificultad para respirar, entonces es probable que tenga ***neuomÃ­a***.

**Regla 2:** Si el usuario tiene tos, sibilancia y opresiÃ³n en el pecho, entonces es probable que tenga ***asma.***

ðŸ“‹ **APLICACIÃ“N EN PHYTON** ðŸ“‹

ðŸ¤” **ExplicaciÃ³n del cÃ³digo del sistema experto para diagnosticar enfermedades respiratorias:**

**FunciÃ³n** ```diagnosticar(sintomas)```**:**

* **ParÃ¡metros:**
    
    ```sÃ­ntomas:``` Lista de sÃ­ntomas del paciente.

* **Retorno:**
    
    La enfermedad mÃ¡s probable que tenga el paciente, o ```None``` si no se puede determinar un diagnÃ³stico.

* **ParÃ¡metros:**
    
    ```enfermedades:``` Diccionario que contiene informaciÃ³n sobre las diferentes enfermedades respiratorias, sus sÃ­ntomas y sus causas.


````python
enfermedades = {
  "neumonia": {
    "sintomas": ["fiebre", "tos", "dificultad para respirar", "taquipnea", "hipotensiÃ³n", "baja saturaciÃ³n de oxiÃ­geno"],
    "causas": ["bacterias", "virus"],
    "tratamiento": ["antibiÃ³ticos", "antivirales"]
  },
  "asma": {
    "sintomas": ["tos", "sibilancias", "opresiÃ³n en el pecho", "falta de aire"],
    "causas": ["alergias", "ejercicio", "aire frÃ­o", "estrÃ©s emocional"],
    "tratamiento": ["broncodilatadores", "esteroides"]
  }
}

def diagnosticar(sintomas):
  diagnostico = None
  for enfermedad, datos in enfermedades.items():
    if all(sintoma in sintomas for sintoma in datos["sintomas"]):
      diagnostico = enfermedad
      break

  return diagnostico

# Ejemplo de uso
sintomas = ["tos", "sibilancias", "opresiÃ³n en el pecho", "falta de aire"]

diagnostico = diagnosticar(sintomas)

print("DiagnÃ³stico:", diagnostico)
````

**Algoritmo:**

1. Se recorre el diccionario ```enfermedades.```
2. Para cada enfermedad, se verifica si todos sus sÃ­ntomas estÃ¡n presentes en la lista de sÃ­ntomas del paciente.
3. Si todos los sÃ­ntomas de una enfermedad estÃ¡n presentes, se retorna esa enfermedad como el diagnÃ³stico.
4. Si no se encuentra ninguna enfermedad con todos sus sÃ­ntomas presentes, se retorna ```None```.


âš¡**ExplicaciÃ³n:**

* La lista ```sintomas``` contiene los sÃ­ntomas "tos", "sibilancia" y "opresiÃ³n en el pecho"

* La funciÃ³n ```diagnosticar()``` recorre el diccionario de ```enfermedades```.

* La funciÃ³n encuentra que la enfermedad "asma" tiene todos sus sÃ­ntomas presentes en la lista ```sintomas```.

* La funciÃ³n retorna "asma" como diagnÃ³stico.
