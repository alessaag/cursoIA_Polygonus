##✅**Caso 1. Sistema de recomendación de Películas**
---



* Un sistema de recomendación de películas podría utilizar el álgebra booleana para recomendar películas a un usuario que ha indicado que **le gustan las películas de acción y ciencia ficción.** El sistema podría buscar películas que sean tanto de acción como de ciencia ficción, y luego recomendarlas al usuario.

⚡**Explicación:**

* La función ```recomendar_peliculas``` recibe un diccionario con las preferencias del usuario como entrada.

* La variable ```peliculas``` es un diccionario que mapea cada género con una lista de películas de ese género.

* La función recorre los géneros favoritos del usuario y agrega las películas de cada género a la lista de recomendaciones.

* Finalmente, se eliminan los duplicados de la lista y se retorna la lista de películas recomendadas.

📋 **APLICACIÓN EN PHYTON** 📋

````python
def recomendar_peliculas(usuario):
  """
  Función que recomienda películas a un usuario en función de sus preferencias.

  Parámetros:
    usuario: Diccionario con las preferencias del usuario.
  Retorno:
    Lista de películas recomendadas.
  """

  peliculas = {
    "accion": ["Duna", "Demon Slayer", "El origen", "Máx", "Resistencia"],#
    "ciencia_ficcion": ["El origen", "La llegada", "Interestelar","The Cloverfield Paradox", "Time Trap"],
    "comedia": ["Wonka", "Ascenso", "Intensamente", "Pared con Pared", "A través de tu Mirada"],
  }

  recomendaciones = []
  for genero in usuario["generos_favoritos"]:
    recomendaciones.extend(peliculas[genero])

  # Eliminar duplicados
  recomendaciones = set(recomendaciones)

  return list(recomendaciones)

usuario = {
  "nombre": "Diego",
  "generos_favoritos": ["accion", "ciencia_ficcion"]
}

recomendaciones = recomendar_peliculas(usuario)

print("Películas recomendadas para", usuario["nombre"], ":", recomendaciones)

````

🤔 **¿Dónde se aplica el álgebra booleana en el caso anterior?**

En el ejemplo anterior, el álgebra booleana se aplica en la siguiente parte:

    for genero in usuario["generos_favoritos"]:
    recomendaciones.extend(peliculas[genero])

En esta sección, se recorre la lista de géneros favoritos del usuario (```usuario["generos_favoritos"]```). Para cada género, se utiliza la clave del género (```genero```) como índice para acceder a la lista de películas de ese género en el diccionario ```peliculas```.

**Aquí es donde se aplica el álgebra booleana:**

* La clave del género (```genero```) se utiliza como condición booleana.
* Si el género está presente en la lista de géneros favoritos del usuario, la expresión se evalúa como ```True```.
* Si el género no está presente en la lista de géneros favoritos del usuario, la expresión se evalúa como ```False```.

##✅**Caso 2. Clasificación de textos y análisis de sentimientos**
---



* Un sistema de análisis de opiniones podría utilizar el álgebra booleana para determinar si un tweet es **positivo**, **negativo o neutral** sobre un producto en particular. El sistema podría buscar palabras clave que se asocian con opiniones positivas o negativas, y luego determinar la opinión general del tweet.

⚡**Explicación:**

* La función ```analizar_opinion``` recibe el texto del tweet y el nombre del producto como entrada.

* Las variables ```palabras_positivas``` y ```palabras_negativas``` contienen listas de palabras que se asocian con opiniones positivas y negativas, respectivamente.

* La función recorre las palabras del tweet y aumenta o disminuye la puntuación en función de si la palabra es positiva o negativa.

* Finalmente, se retorna la opinión general del tweet en función de la puntuación:

        Si la puntuación es positiva, se retorna "Positivo".

        Si la puntuación es negativa, se retorna "Negativo".

        Si la puntuación es cero, se retorna "Neutral".

📋 **APLICACIÓN EN PHYTON** 📋

````python
def analizar_opinion(tweet, producto):
  """
  Función que analiza la opinión sobre un producto en un tweet.

  Parámetros:
    tweet: Texto del tweet.
    producto: Nombre del producto.

  Retorno:
    "Positivo", "Negativo" o "Neutral".
  """

  palabras_positivas = ["bueno", "excelente", "maravilloso", "me gusta", "me encanta", "lo quiero", "lo necesito"]
  palabras_negativas = ["malo", "terrible", "horrible", "no me gusta", "pésimo", "desagradable", "mal estado"]

  puntuacion = 0
  for palabra in tweet.split():
    if palabra in palabras_positivas:
        puntuacion += 1
    elif palabra in palabras_negativas:
      puntuacion -= 1

  if puntuacion > 0:
    return "Positivo"
  elif puntuacion < 0:
    return "Negativo"
  else:
    return "Neutral"

tweet = "El nuevo teléfono #SamsungGalaxyS23 es excelente! #Tecnologia #Smartphone"
producto = "Samsung Galaxy S23"

import re
#Utilizamos una expresión regular para eliminar los símbolos
texto_limpio = re.sub(r'[^\w\s]', '', tweet)

opinion = analizar_opinion(texto_limpio, producto)

print("Opinión sobre", producto, ":", opinion)
````

**1. Base de conocimiento:**

**2. Recopilación de síntomas**

* Se le pregunta al usuario sobre los síntomas que está experimentando.

**3. Análisis de síntomas**

* Se utiliza la información de la base de conocimiento y el álgebra booleana para analizar los síntomas del usuario y determinar la enfermedad más probable.

* Se pueden utilizar reglas booleanas para determinar si un conjunto de síntomas es compatible con una enfermedad en particular.


**4. Presentación del diagnóstico:**

* Se presenta al usuario el diagnóstico más probable, junto con información adicional sobre la enfermedad, como sus causas, tratamiento y prevención.

🦠**EJEMPLOS DE APLICACIÓN DEL ÁLGEBRA BOOLEANA** 🦠

**Regla 1:** Si el usuario tiene fiebre, tos y dificultad para respirar, entonces es probable que tenga ***neuomía***.

**Regla 2:** Si el usuario tiene tos, sibilancia y opresión en el pecho, entonces es probable que tenga ***asma.***

📋 **APLICACIÓN EN PHYTON** 📋

🤔 **Explicación del código del sistema experto para diagnosticar enfermedades respiratorias:**

**Función** ```diagnosticar(sintomas)```**:**

* **Parámetros:**
    
    ```síntomas:``` Lista de síntomas del paciente.

* **Retorno:**
    
    La enfermedad más probable que tenga el paciente, o ```None``` si no se puede determinar un diagnóstico.

* **Parámetros:**
    
    ```enfermedades:``` Diccionario que contiene información sobre las diferentes enfermedades respiratorias, sus síntomas y sus causas.


````python
enfermedades = {
  "neumonia": {
    "sintomas": ["fiebre", "tos", "dificultad para respirar", "taquipnea", "hipotensión", "baja saturación de oxiígeno"],
    "causas": ["bacterias", "virus"],
    "tratamiento": ["antibióticos", "antivirales"]
  },
  "asma": {
    "sintomas": ["tos", "sibilancias", "opresión en el pecho", "falta de aire"],
    "causas": ["alergias", "ejercicio", "aire frío", "estrés emocional"],
    "tratamiento": ["broncodilatadores", "esteroides"]
  }
}

def diagnosticar(sintomas):
  diagnostico = None
  for enfermedad, datos in enfermedades.items():
    if all(sintoma in sintomas for sintoma in datos["sintomas"]):
      diagnostico = enfermedad
      break

  return diagnostico

# Ejemplo de uso
sintomas = ["tos", "sibilancias", "opresión en el pecho", "falta de aire"]

diagnostico = diagnosticar(sintomas)

print("Diagnóstico:", diagnostico)
````

**Algoritmo:**

1. Se recorre el diccionario ```enfermedades.```
2. Para cada enfermedad, se verifica si todos sus síntomas están presentes en la lista de síntomas del paciente.
3. Si todos los síntomas de una enfermedad están presentes, se retorna esa enfermedad como el diagnóstico.
4. Si no se encuentra ninguna enfermedad con todos sus síntomas presentes, se retorna ```None```.


⚡**Explicación:**

* La lista ```sintomas``` contiene los síntomas "tos", "sibilancia" y "opresión en el pecho"

* La función ```diagnosticar()``` recorre el diccionario de ```enfermedades```.

* La función encuentra que la enfermedad "asma" tiene todos sus síntomas presentes en la lista ```sintomas```.

* La función retorna "asma" como diagnóstico.
