**1. Definición formal:**

Los ***teoremas de Boole*** son dos teoremas fundamentales del álgebra de Boole que establecen las propiedades básicas de las operaciones lógicas **AND, OR y NOT.**

✅**Teorema de la identidad**
---



Establece que para cualquier proposición P:

    P or P = P
    P and P = P

⚡**En palabras simples:**

* Si una proposición es verdadera, al combinarla con ella misma sigue siendo verdadera.
* Si una proposición es falsa, al combinarla con ella misma sigue siendo falsa.

🤔**Ejemplo:**

▶ Si **P** es la proposición "Está lloviendo", entonces **P and P** es la proposición "Está lloviendo y está lloviendo". Ambas proposiciones son equivalentes, ya que SI está lloviendo, NO hay posibilidad de que no esté lloviendo al mismo tiempo.

✅**Explicación:**

* La función ```es_mamifero``` define una regla que determina si un animal es mamífero.

* La regla se basa en el teorema de la identidad de Boole: si un animal tiene pelo y cuatro patas (ambas condiciones son necesarias), entonces es un mamífero.

````python
def es_mamifero(pelo, cuatro_patas):
  """
  Determina si un animal es un mamífero.

  Args:
    pelo: True si el animal tiene pelo, False si no.
    cuatro_patas: True si el animal tiene cuatro patas, False si no.

  Returns:
    True si el animal es un mamífero, False si no.
  """
  # Regla 1: Si un animal tiene pelo y cuatro patas, entonces es un mamífero.
  return pelo and cuatro_patas

# Ejemplo de uso

animal1 = {"pelo": True, "cuatro_patas": True}  # Si .....
animal2 = {"pelo": False, "cuatro_patas": True}  # No ....
animal3 = {"pelo": True, "cuatro_patas": False}  # No .....

print(f"Animal 1: ¿Es mamífero? {es_mamifero(**animal1)}")  # Output: Animal 1: Mamífero? True
print(f"Animal 2: ¿Es mamífero? {es_mamifero(**animal2)}")  # Output: Animal 2: Mamífero? False
print(f"Animal 3: ¿Es mamífero? {es_mamifero(**animal3)}")  # Output: Animal 3: Mamífero? False

````

✅**Teorema del complemento**
---

Estable que para cualquier proposición P:

        P and not P = False
        P or not P  = True



⚡**En palabras simples:**

* Una proposición y su negación no pueden ser ambas verdaderas al mismo tiempo.
* Si una proposición es verdadera, su negación es falsa.
* Si una proposición es falsa, su negación es verdadera.

🤔**Ejemplo:**

▶ Si **P** es la proposición "El cielo es azul", entonces **not P** es la proposición "El cielo NO es azul". Ambas proposiciones no pueden ser verdaderas al mismo tiempo, ya que el cielo no puede ser azul y no ser azul al mismo tiempo.

⚡**Explicación:**

* La función ```filtrar_datos``` toma dos argumentos: una lista de datos y una función de condición.

* La función de condición recibe un diccionario de datos como argumento y devuelve un valor booleano.

* La función itera sobre la lista de datos y agrega cada diccionario a la lista de datos filtrados si la función devuelve ```True``` para ese diccionario.

**Teorema del complemento de Boole**

✅ La función de condición devuelve ```True``` si la edad de la persona es mayor o igual a 21.

✅ La función de condición devuelve ```False``` si la edad de la persona es menor a 21.

**Filtrar datos:**

````python
def filtrar_datos(datos, condicion):
  """
  Filtra un conjunto de datos según una condición booleana.

  Args:
    datos: Lista de diccionarios con datos.
    condicion: Función que recibe un diccionario de datos y devuelve un valor booleano.

  Returns:
    Lista de diccionarios de datos que cumplen la condición.
  """
  datos_filtrados = []
  for dato in datos:
    if condicion(dato):
      datos_filtrados.append(dato)
  return datos_filtrados

# Ejemplo de uso
datos = [
  {"nombre": "Juan", "edad": 20},
  {"nombre": "María", "edad": 30},
  {"nombre": "Pedro", "edad": 18},
  {"nombre": "Ana", "edad": 25},
]

condicion = lambda dato: dato["edad"] >= 20

datos_filtrados = filtrar_datos(datos, condicion)

print(f"Datos filtrados: {datos_filtrados}")
````

**Leyes de Morgan en palabras simples**

* Son dos reglas que nos permiten transformar expresiones lógicas que incluyen negaciones. Son una herramienta útil para simplificar y comprender mejor el significado de estas expresiones.

✅ **Primera ley de Morgan:**
---


    ¬(P ∧ Q) = ¬P ∨ ¬Q

En palabras simples, esta ley dice que la negación de una conjunción ("y") es equivalente a la disyunción ("o") de las negaciones.

🔥**Ejemplo:**

* Si la proposición **P** es "Está soleado" y la proposición **Q** es "Hace calor", entonces la expresión "No está soleado y no hace calor" es equivalente a "No está soleado o no hace calor".

✅**Segunda ley de Morgan:**
---


    ¬(P ∨ Q) = ¬P ∧ ¬Q

Esta ley dice que la negación de una disyunción ("o") es equivalente a la conjunción ("y") de las negaciones.

🔥**Ejemplo:**

Si la proposición **P** es "Es lunes" y la proposición **Q** es "Es feriado", entonces la expresión "No es lunes o no es feriado" es equivalente a "No es lunes y no es feriado"

**1. Clasificación de texto:**

**Problema:**

Clasificar un correo electrónico como spam o no spam.

**Solución: Definir dos características**

**P:** Contiene palabras clave de spam.

**Q:** Contiene palabras clave legítimas.

````python
#Usar las leyes de Morgan para crear una regla de clasificación:

def es_spam(texto):
  """
  Determina si un texto es spam.

  Args:
    texto: El texto a analizar.

  Returns:
    True si el texto es spam, False si no lo es.
  """
  palabras_spam = ["oferta", "gratis", "dinero", "ahora"]
  palabras_legitimas = ["hola", "gracias", "información", "contacto"]

  p = any(palabra in texto for palabra in palabras_spam)
  q = any(palabra in texto for palabra in palabras_legitimas)

  # Ley de Morgan: negar la disyunción
  return not (p and q)

# Ejemplo de uso
texto = "Oferta especial, ¡obtenga dinero gratis ahora!"

es_spam = es_spam(texto)

if es_spam:
  print("El texto es spam.")
else:
  print("El texto no es spam.")

````

⚡**Definición**

* La función ```es_vertebrado``` define una regla que determina si un animal es vertebrado.

* Crear una regla que se  base en el teorema de la identidad de Boole: si un animal tiene ciertas carácterísticas (Animal vertebrado), entonces en un animal vertebrado.

* Recomendaciones: "```Ver ejemplo de mamífero```"

````python
def es_vertebrado(simetria_bilateral, columna_vertebral):
  """
  Determina si un animal es un vertebrado.

  Args:
    simetria_bilateral: True si el animal tiene dos lados del cuerpo: izquierdo y derecho fácilmente detectables, False si no.
    columna vertebral: True si el animal tiene vértebras, False si no.

  Returns:
    True si el animal es un vertebrado, False si no.
  """
  # Regla 1: Si un animal tiene simetria_bilateral y columna_vertebral, entonces es un vertebrado.
  return simetria_bilateral and columna_vertebral

# Ejemplo de uso

animal1 = {"simetria_bilateral": True, "columna_vertebral": True}  # Si .....
animal2 = {"simetria_bilateral": False, "columna_vertebral": True}  # No ....
animal3 = {"simetria_bilateral": True, "columna_vertebral": False}  # No .....

print(f"Animal 1: ¿Es vertebrado? {es_vertebrado(**animal1)}")  # Output: Animal 1: vertebrado? True
print(f"Animal 2: ¿Es vertebrado? {es_vertebrado(**animal2)}")  # Output: Animal 2: vertebrado? False
print(f"Animal 3: ¿Es vertebrado? {es_vertebrado(**animal3)}")  # Output: Animal 3: vertebrado? False
````

⚡**Problema:**

Filtrar tweets que contengan información relevante sobre un tema específico.

**Solución:**

Definir dos palabras clave:

* **Palabra_clave_1:** Tema principal del tweet.
* **Palabra_clave_2:** Tema secundario del tweet.

````python
#Usar las leyes de Morgan para crear una regla de clasificación:

def tema_tweet(texto):
  """
  Determina cual es el tema de un tweet.

  Args:
    texto: El texto a analizar.

  Returns:
    True si el texto es política, False si el texto es religión.
  """
  palabras_politica = ["país", "vota", "elige", "cambio"]
  palabras_religion = ["Dios", "oración", "fe", "Jesús"]

  p = any(palabra in texto for palabra in palabras_politica)
  r = any(palabra in texto for palabra in palabras_religion)

  # Ley de Morgan: negar la disyunción
  return not (p and r)

# Ejemplo de uso
texto1 = "Dios es amor, expresa tu fe a través de la oración"
texto2 = "Por el cambio del país elige Fajardo"

tema_tweet = tema_tweet(texto2)

if tema_tweet:
  print("El texto es política.")
else:
  print("El texto es religión.")
````
