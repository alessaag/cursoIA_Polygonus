# 游꺔 **<font color='#AECF00'>Medidas de Tendencias Central:</font>**

```python
import numpy as np

# Datos de ejemplo
calificaciones = [70, 80, 85, 90, 95, 70, 82]

# C치lculo de medidas de tendencia central
media = np.mean(calificaciones)
mediana = np.median(calificaciones)
moda = np.argmax(np.bincount(calificaciones))

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)

print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
```

# 游꺔 **<font color='#AECF00'>Medidas de tendencias dispersi칩n</font>**

```python
# C치lculo de medidas de dispersi칩n
varianza = np.var(calificaciones)
desviacion_estandar = np.std(calificaciones)

print("Varianza:", varianza)
print("Desviaci칩n Est치ndar:", desviacion_estandar)

print(f"Varianza: {varianza:.2f}")
print(f"Desviaci칩n Est치ndar: {desviacion_estandar:.2f}")
```

# 游꺔 **<font color='#AECF00'>Introducci칩n a la distribuci칩n normal:</font>**

```python
# Importamos las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci칩n de estilo para gr치ficos
sns.set(style="whitegrid")

# Generamos datos simulados que siguen una distribuci칩n normal
data = np.random.normal(loc=50, scale=10, size=1000)

# C치lculo de la media y la mediana
mean_value = np.mean(data)
median_value = np.median(data)

# Visualizaci칩n de la distribuci칩n normal con l칤neas de la media y mediana
plt.figure(figsize=(10,6))
sns.histplot(data, bins=30, kde=True, color='blue')

# A침adimos las l칤neas de la media y la mediana
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'Media: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='-', linewidth=2, label=f'Mediana: {median_value:.2f}')

# T칤tulos y etiquetas
plt.title('Distribuci칩n normal de los datos', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Mostramos la leyenda
plt.legend()

# Mostramos el gr치fico
plt.show()

# Imprimimos la media y la mediana en la consola
print(f"Media de los datos: {mean_value:.2f}")
print(f"Mediana de los datos: {median_value:.2f}")
```
