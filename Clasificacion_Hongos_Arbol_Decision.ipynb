```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Importar dataset, tomado de: https://www.kaggle.com/datasets/mehmetisik/mushrooms/data
df = pd.read_csv('mushrooms.csv')

# Visualización
print(df.head())
```

```python
# Visualización

# 1. Distribución de Clases
plt.figure(figsize=(8, 6))
sns.countplot(x='class', data=df)
plt.title('Distribución de Hongos comestibles (e) y venenosos (p)')
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# 2. Cap Shape vs. Class
plt.figure(figsize=(10, 6))
sns.countplot(x='cap-shape', hue='class', data=df)
plt.title('Distribución de Forma de Cabeza por Clase')
plt.xlabel('Forma de Cabeza')
plt.ylabel('Frecuencia')
plt.legend(title='Clase', loc='upper right')
plt.show()
```

```python
# 3. Cap Color vs. Class
plt.figure(figsize=(10, 6))
sns.countplot(x='cap-color', hue='class', data=df)
plt.title('Distribución de Color de Cabeza por Clase')
plt.xlabel('Color de Cabeza')
plt.ylabel('Frecuencia')
plt.legend(title='Clase', loc='upper right')
plt.show()
```

```python
# 4. Odor vs. Class
plt.figure(figsize=(10, 6))
sns.countplot(x='odor', hue='class', data=df)
plt.title('Distribución de Olor por Clase')
plt.xlabel('Olor')
plt.ylabel('Frecuencia')
plt.legend(title='Clase', loc='upper right')
plt.show()
```

```python
# # Dataset Features:

##** Class**: Indicates if the mushroom is poisonous (p) or edible (e).

# Cap Shape: Describes the shape of the mushroom cap (e.g., x: convex, c: conical, f: flat, k: knobbed, s: sunken, b: bell-shaped).

# Cap Surface: Texture of the mushroom cap's surface (e.g., f: fibrous, g: grooves, y: scaly, s: smooth).

# Cap Color: Color of the cap (e.g., n: brown, b: buff, c: cinnamon, g: gray, r: green, p: pink, u: purple, e: red, w: white y: yellow).

# Bruises: Whether the mushroom has bruises (t: yes, f: no).

# Odor: Scent of the mushroom (e.g., p: pungent, a: almond, l: anise, c: creosote, y: fishy, f: foul, m: musty, n: none, s: spicy).

# Gill Attachment: How the gills are attached to the stalk (f: free, a: attached, d: descending, n: notched).

# Gill Spacing: Spacing between the gills (c: close, w: crowded, d: distant).

# Gill Size: Size of the gills (b: broad, n: narrow).

# Gill Color: Color of the gills (black=k, brown=n, buff=b, chocolate=h, gray=g, green=r, orange=o, pink=p, purple=u, red=e, white=w).

# Stalk Shape: Shape of the stalk (e: enlarging, t: tapering).

# Stalk Root: Type of the stalk root (e.g., b: bulbous, club=c, cup=u, equal=e, rhizomorphs=z, rooted=r, missing=?).

# Stalk Surface Above Ring: Texture of the stalk surface above the ring (fibrous=f, scaly=y, silky=k, smooth=s).

# Stalk Surface Below Ring: Texture of the stalk surface below the ring (fibrous=f, scaly=y, silky=k, smooth=s).

# Stalk Color Above Ring: Color of the stalk above the ring.

# Stalk Color Below Ring: Color of the stalk below the ring.

# Veil Type: Type of veil (p: partial).

# Veil Color: Color of the veil.

# Ring Number: Number of rings on the stalk (o: one, t: two).

# Ring Type: Type of ring (e.g., p: pendant).

# Spore Print Color: Color of the mushroom's spore print.

# Population: Density of the mushroom population (e.g., s: scattered).

#** Habitat**: Habitat where the mushroom is found (e.g., u: urban, g: grassland).
```
